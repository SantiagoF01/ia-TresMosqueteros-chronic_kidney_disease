# 1. Problema y Datos

El dataset Chronic Kidney Disease (CKD) contiene 400 registros de pacientes con 28 variables clínicas, incluyendo edad, presión arterial y resultados de laboratorio. El objetivo es clasificar el estado de la enfermedad renal crónica a partir de estas características. Este problema es relevante debido a la importancia de la detección temprana de la enfermedad, y presenta desafíos como la presencia de datos heterogéneos y valores faltantes.

# 2. Justificación de la IA

Dado que las relaciones entre las variables clínicas y la enfermedad renal crónica no suelen ser lineales ni fácilmente identificables mediante análisis tradicionales, el uso de técnicas de Inteligencia Artificial supervisada se presenta como una alternativa robusta para el análisis de este tipo de datos. Estas técnicas permiten modelar interacciones complejas entre múltiples variables, identificando patrones ocultos que no son evidentes a simple vista.

Además, los modelos supervisados pueden aprender a partir de datos previamente etiquetados, lo que facilita la construcción de sistemas capaces de clasificar con mayor precisión la presencia y el grado de la enfermedad. Esto resulta especialmente relevante en el ámbito de la salud, donde una detección temprana y precisa puede impactar significativamente en el tratamiento y la calidad de vida de los pacientes.

En comparación con enfoques tradicionales basados en reglas o umbrales fijos, la Inteligencia Artificial ofrece una mayor capacidad de adaptación a la variabilidad de los datos clínicos, reduciendo errores y mejorando la automatización de procesos diagnósticos, lo cual puede servir como apoyo a la toma de decisiones médicas.

## 1. Filtración glomerular

```python
# Númerica Asimétrica
# Revisar correlación
print("Correlacion con gfr: ", dataset['gfr'].corr(dataset['ckd_stage']))

# Filtar nulos
df = dataset.dropna(subset=['gfr'])

# Estadísticas
media = df['gfr'].mean()
desv = df['gfr'].std()
mediana = df['gfr'].median()
q1 = df['gfr'].quantile(0.25)
q3 = df['gfr'].quantile(0.75)
print("Promedio: ", media)
print("Desviación: ", desv)
print("Mediana: ", mediana)
print("IQR: ", (q3 - q1))

# Crear figura
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Distribución de GFR')

# Crear histograma
sns.histplot(data=df, x='gfr', bins=50, kde=True, ax=axes[0], color='steelblue')
axes[0].axvline(mediana, color='red', linestyle='--', linewidth=1.5, label=f'Mediana: {mediana:.3f}')
axes[0].set_title('Histograma con KDE')
axes[0].set_xlabel('GFR')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()

# Crear boxplot
sns.boxplot(data=df, x='gfr', ax=axes[1], color='steelblue')
axes[1].axvline(mediana, color='red', linestyle='--', linewidth=1.5, label=f'Mediana: {mediana:.3f}')
axes[1].set_title('Boxplot')
axes[1].set_xlabel('GFR')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Seleccion de variables
features = ['packed_cell_volume', 'hemoglobin', 'red_blood_cell_count', 'blood_urea', 'serum_creatinine', 'sodium']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(features):
  # Grafica base - Seaborn
  sns.histplot(df[col], kde=True, ax=axes[i], color='mediumseagreen', bins=30)

  # Refinamiento - Matplotlib
  axes[i].set_title(f'Distribución de {col.capitalize()}', fontweight='bold')
  axes[i].set_xlabel('')
  axes[i].set_ylabel('Frecuencia')
  sns.despine(ax=axes[i])

  # Revisión de simetría
  skewness = df[col].skew()

  if abs(skewness) < 0.5:
    # Numérica Simétrica -> Media y Desviación
    mean = df[col].mean()
    std = df[col].std()
    axes[i].axvline(mean, color='navy', linestyle='--', linewidth=1.5, label=f'Mean: {mean:.2f}')
    axes[i].legend()
    print(f"[{col}] Distribución Simétrica (sesgo: {skewness:.2f}).")
    print(f"Reporte: El valor promedio de {col} es de ${mean:.2f} \pm {std:.2f}. $ \n")
  else:
    # Numérica Asimétrica -> Mediana y IQR
    median = df[col].median()
    q3 = df[col].quantile(0.75)
    q1 = df[col].quantile(0.25)
    IQR = q3 - q1
    axes[i].axvline(median, color='navy', linestyle='--', linewidth=1.5, label=f'Median: {median:.2f}')
    axes[i].legend()
    print(f"[{col}] Distribución Asimétrica (sesgo: {skewness:.2f}).")
    print(f"Reporte: El {col} tiene una mediana de {median:.2f} y un rango intercuartil (IQR) de {IQR:.2f}. \n")

plt.tight_layout()
plt.show()
```
